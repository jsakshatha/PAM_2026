# 1️⃣ INTRODUCTION TO ANN

Artificial Neural Network (ANN) is a machine learning model inspired by the human brain.

ANN consists of neurons (processing units).

Each neuron:
- Takes inputs
- Multiplies inputs with weights
- Adds bias
- Applies activation function
- Produces output

ANN is mainly used for:
- Image classification
- Prediction
- Pattern recognition
- Speech recognition

---

# 2️⃣ BASIC STRUCTURE OF ANN

ANN has three layers:

1. Input Layer  
2. Hidden Layer  
3. Output Layer  

Simple structure:

3 Inputs → 1 Neuron → 1 Output

# 4️⃣ MATHEMATICAL REPRESENTATION

Weighted Sum:

$$
Z = X_1W_1 + X_2W_2 + X_3W_3 + b
$$

Activation Function:

$$
Output = f(Z)
$$

If Sigmoid is used:

$$
f(Z) = \frac{1}{1 + e^{-Z}}
$$

---

# 5️⃣ MEANING OF SYMBOLS (TABLE FORMAT)

| Symbol | Name | Meaning | Mathematical Role |
|--------|------|----------|-------------------|
| $$X_i$$ | Input | Data entering neuron | Independent variable |
| $$W_i$$ | Weight | Importance of input | Multiplier |
| $$b$$ | Bias | Adjustment term | Constant shift |
| $$Z$$ | Weighted Sum | Total before activation | Linear combination |
| $$f(Z)$$ | Activation | Final output | Non-linear transformation |

# 6️⃣ COMPLETE NUMERICAL EXAMPLE (ONE DATA SET)

Given:

$$
X_1 = 2
$$

$$
X_2 = 1
$$

$$
X_3 = 3
$$

$$
W_1 = 0.5
$$

$$
W_2 = -1
$$

$$
W_3 = 0.8
$$

$$
b = 0.2
$$

---

## Step 1: Calculate Z

$$
Z = (2 \times 0.5) + (1 \times -1) + (3 \times 0.8) + 0.2
$$

$$
Z = 1 - 1 + 2.4 + 0.2
$$

$$
Z = 2.6
$$

---

## Step 2: Apply Sigmoid Activation

$$
f(Z) = \frac{1}{1 + e^{-2.6}}
$$

$$
e^{-2.6} \approx 0.074
$$

$$
f(Z) = \frac{1}{1 + 0.074}
$$

$$
f(Z) = \frac{1}{1.074}
$$

$$
f(Z) \approx 0.93
$$

---

# 7️⃣ FINAL OUTPUT INTERPRETATION

Decision Rule:

$$
\text{If } f(Z) > 0.5 \rightarrow Class\ 1
$$

$$
\text{If } f(Z) < 0.5 \rightarrow Class\ 0
$$

Here:

$$
f(Z) \approx 0.93
$$

Therefore:

Class 1

---

# 8️⃣ GENERAL FORM (FOR N INPUTS)

For n inputs:

$$
Z = \sum_{i=1}^{n} X_i W_i + b
$$

Output:

$$
y = f(Z)
$$

# 🔷 STEP 1: Input → Hidden Layer

## 🔹 Hidden Neuron $$h_1$$

### Linear Combination

$$
z_1 = (x_1 w_{11}) + (x_2 w_{21}) + b_1
$$

$$
z_1 = (1 \times 0.5) + (2 \times 0.3) + 0.1
$$

$$
z_1 = 0.5 + 0.6 + 0.1
$$

$$
z_1 = 1.2
$$

### Sigmoid Activation

$$
h_1 = \frac{1}{1 + e^{-z_1}}
$$

$$
h_1 = \frac{1}{1 + e^{-1.2}}
$$

$$
e^{-1.2} \approx 0.301
$$

$$
h_1 = \frac{1}{1.301}
$$

$$
h_1 \approx 0.768
$$

---

## 🔹 Hidden Neuron $$h_2$$

### Linear Combination

$$
z_2 = (x_1 w_{12}) + (x_2 w_{22}) + b_2
$$

$$
z_2 = (1 \times 0.4) + (2 \times 0.2) + 0.2
$$

$$
z_2 = 0.4 + 0.4 + 0.2
$$

$$
z_2 = 1.0
$$

### Sigmoid Activation

$$
h_2 = \frac{1}{1 + e^{-1}}
$$

$$
e^{-1} \approx 0.368
$$

$$
h_2 = \frac{1}{1.368}
$$

$$
h_2 \approx 0.731
$$

---

# 🔷 STEP 2: Hidden → Output Layer (Regression)

⚠️ For Regression → No Sigmoid at Output  
We use Linear Activation.

### Linear Combination

$$
z_o = (h_1 v_1) + (h_2 v_2) + b_o
$$

$$
z_o = (0.768 \times 0.6) + (0.731 \times 0.7) + 0.3
$$

$$
z_o = 0.4608 + 0.5117 + 0.3
$$

$$
z_o = 1.2725
$$

---

## Final Output

$$
y = z_o
$$

$$
\boxed{y \approx 1.2725}
$$
